```python
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
```

```python
test_x = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTest_x.csv', header = None)
train_x = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTrain_x.csv', header = None)
train_y = pd.read_csv('../input/2022-ml-finalexam-p2/NetworkTrain_y.csv', header = None)
submit = pd.read_csv('../input/2022-ml-finalexam-p2/submission_sample.csv', header = 0)
```

```python
train_x
```

```python
# 1번 column의 데이터 맵핑

train_x[1][train_x[1] == 'tcp'] = 0
train_x[1][train_x[1] == 'udp'] = 1
train_x[1][train_x[1] == 'icmp'] = 2
train_x[1] = train_x[1].astype('int64')


test_x[1][test_x[1] == 'tcp'] = 0
test_x[1][test_x[1] == 'udp'] = 1
test_x[1][test_x[1] == 'icmp'] = 2
test_x[1] = test_x[1].astype('int64')
```

```python
pd.DataFrame(train_x)
```

```python
# 2번 column의 데이터 실수화
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_x[2] = le.fit_transform(train_x[2])
test_x[2] = le.transform(test_x[2])
```

```python
# 2번 column의 데이터 실수화
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_x[3] = le.fit_transform(train_x[3])
test_x[3] = le.transform(test_x[3])
```

```python
# 전처리가 잘 됐는지 확인
pd.DataFrame(train_x)
```

```python
# 데이터 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
# from sklearn.decomposition import PCA #-> 성능 저하
# pca = PCA()
# train_x = pca.fit_transform(train_x)
# test_x = pca.transform(test_x)
```

```python
# randomforestclassifier를 통하여 범주형 라벨 예측
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = seed, n_estimators = 25)
train_y = np.array(train_y).flatten()
rfc.fit(train_x, train_y)
pred = rfc.predict(test_x)
```

```python
# 현재 상태에 이상이 없다면(normal) 해당 pred 값을 0으로, 문제가 있다면 1로
for i in range(len(pred)):
    if pred[i] == 'normal':
        pred[i] = 0
    else :pred[i] = 1
```

```python
submit['Category'] = pred
```

```python
submit.to_csv('submission.csv', index = False)
```
