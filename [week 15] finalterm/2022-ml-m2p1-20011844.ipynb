```python
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)
```

```python
submit = pd.read_csv('../input/2022-ml-finalexam-p1/sample_submission.csv')
train = pd.read_csv('../input/2022-ml-finalexam-p1/train.csv')
test = pd.read_csv('../input/2022-ml-finalexam-p1/test.csv')
```

```python
# Id, groupId, matchId drop
train = train.drop(['Id', 'groupId', 'matchId'], axis = 1)
test = test.drop(['Id', 'groupId', 'matchId'], axis = 1)
```

```python
# trainX, trainYë¡œ ë¶„ë¦¬
trainX = train.drop(['winPlacePerc'], axis = 1)
trainY = train['winPlacePerc']
```

```python
# randomforestregressorë¥¼ í†µí•˜ì—¬ íšŒê·€ ì˜ˆì¸¡
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = seed)
# í•™ìŠµ
rfr.fit(trainX, trainY)
# ì˜ˆì¸¡
pred = rfr.predict(test)
```

```python
submit['winPlacePerc'] = pred
submit.to_csv('submission.csv', index = False)
```
