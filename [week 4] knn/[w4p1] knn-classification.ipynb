```python
# 데이터셋 불러오기
train = pd.read_csv('/kaggle/input/2022-ml-w4p1/train.csv')
sample = pd.read_csv('/kaggle/input/2022-ml-w4p1/sample.csv')
test = pd.read_csv('/kaggle/input/2022-ml-w4p1/test.csv')
print(test)
```

```python
x = train.drop('label', axis = 1)
y = train['label']
print(x.head())
print(y.head())
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1, stratify = y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x_sc = scaler.transform(x) # train data (x) 표준화
test_sc = scaler.transform(test) # test data 표준화

#표준화 안 한 게 성능 더 good
```

```python
# 모형 추정 및 사례중심 학습
from sklearn.neighbors import KNeighborsClassifier #KNN 불러오기
knn = KNeighborsClassifier(n_neighbors = 3, p = 1) # 3개의 인접한 이웃, 거리측정기준: 유클리드
knn.fit(x_train, y_train)
```

```python
#결과 분석
y_train_pred = knn.predict(x_train)
y_test_pred = knn.predict(x_test)
y_pred = knn.predict(x)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_test_pred))
print(accuracy_score(y, y_pred))
```
